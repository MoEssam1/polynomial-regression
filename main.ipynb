# Importing the librares

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

# Import dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# Visualising the dataset

```python
plt.scatter(x,y,color = "red")
plt.title("visualising dataset")
plt.xlabel("Level")
plt.ylabel("salary")
plt.show()
```

# Train the poly regression model on the whole dataset 

```python
from sklearn.preprocessing import PolynomialFeatures
polyFeatures = PolynomialFeatures(degree=4)
x_poly = polyFeatures.fit_transform(x)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_poly,y)
```

# Visualising polynomial regression result 

```python
plt.scatter(x,y,color = "red")
plt.plot(x,regressor.predict(x_poly))
plt.title("visualising the poly regression model")
plt.xlabel("Level")
plt.ylabel("salary")
plt.show()
```

## Visualising polynomial regression result (for higher resoultion and smoother curve)

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
xGrid_poly = polyFeatures.fit_transform(x_grid)
plt.scatter(x,y,color = "red")
plt.plot(x_grid,regressor.predict(xGrid_poly))
plt.title("visualising the poly regression model")
plt.xlabel("Level")
plt.ylabel("salary")
plt.show()
```

# Predicting a New result

```python
y_predict = regressor.predict(polyFeatures.fit_transform([[6.5]]))
print(y_predict)
```
